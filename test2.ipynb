```python
import difflib

file1_path = "myMostProbableLocationWithGrid10_tillTime100.csv"
file2_path = "mostProbableLocationWithGrid10_tillTime100.csv"

with open(file1_path, 'r') as f1, open(file2_path, 'r') as f2:
    file1_lines = f1.readlines()
    file2_lines = f2.readlines()

d = difflib.Differ()
diff = d.compare(file1_lines, file2_lines)

print("".join(diff))
```
